```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lifeexpectancy.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
sns.set_theme(style="whitegrid")
plt.style.use("ggplot")
```

```python
df.columns = df.columns.str.strip()  # Remove whitespace from column names
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.columns = df.columns.str.strip().str.replace(' +', ' ', regex=True)
```

```python
for col in ['Population', 'GDP', 'Total expenditure', 'Alcohol', 'BMI',
            'thinness 1-19 years', 'thinness 5-9 years']:
     df[col] = df[col].fillna(df[col].median())
```

```python
print(df.columns.tolist())
```

```python
print(df.columns.tolist())
```

```python
# Fill numeric values with MEDIAN
for col in ['Population', 'GDP', 'Total expenditure', 'Alcohol', 'BMI',
            'thinness 1-19 years', 'thinness 5-9 years']:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# Fill health-related % values with MEDIAN
for col in ['Hepatitis B', 'Polio', 'Diphtheria']:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# Fill socio-economic indicators with MEDIAN
for col in ['Income composition of resources', 'Schooling']:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# Fill target variable: Life expectancy
df['Life expectancy'].fillna(df['Life expectancy'].mean(), inplace=True)
```

```python
# Final check
print(df.isnull().sum().sort_values(ascending=False))
```

```python
df['Adult Mortality'].fillna(df['Adult Mortality'].median(), inplace=True)
```

```python
# Unique values in categorical columns
print(df['Country'].nunique())     # Number of countries
print(df['Status'].value_counts()) # Count of Developing vs Developed
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.histplot(df['Life expectancy'], kde=True, bins=30)
plt.title('Life Expectancy Distribution')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='GDP', y='Life expectancy', hue='Status', data=df)
plt.title('Life Expectancy vs GDP by Status')
plt.show()
```

```python
list(df.groupby('Country')['Life expectancy'])
```

```python
# Mean life expectancy by country
country_life_exp = df.groupby('Country')['Life expectancy'].mean().sort_values()
# Bottom 10
country_life_exp.head(10).plot(kind='barh', title='Bottom 10 Countries by Life Expectancy')
plt.show()
```

```python
# Top 10
country_life_exp.tail(10).plot(kind='barh', title='Top 10 Countries by Life Expectancy', color='green')
plt.show()
```

```python
sns.countplot(x='Status', data=df)
plt.title('Distribution of Development Status')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x='Year', y='Life expectancy', hue='Status', estimator='mean')
plt.title('Average Life Expectancy Over Years')
plt.show()
```

```python
sns.pairplot(df[['Life expectancy', 'BMI']])
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
# Boxplots for important numerical features
num_cols = ['Life expectancy', 'Alcohol', 'BMI', 'GDP', 'Schooling']

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Trend of Schooling and Alcohol consumption
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.lineplot(x='Year', y='Schooling', data=df, estimator='mean', ax=ax[0])
sns.lineplot(x='Year', y='Alcohol', data=df, estimator='mean', ax=ax[1])
ax[0].set_title('Average Schooling Over Years')
ax[1].set_title('Average Alcohol Consumption Over Years')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='Income composition of resources', y='Life expectancy', hue='Status', data=df)
plt.title('Life Expectancy vs Income Composition of Resources')
plt.show()
```

```python
sns.scatterplot(x='Adult Mortality', y='Life expectancy', hue='Status', data=df)
plt.title('Life Expectancy vs Adult Mortality')
plt.show()
```
